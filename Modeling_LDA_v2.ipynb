### LDA on words without lammetization - 10 & 20 topics

```python
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.decomposition import TruncatedSVD
from sklearn.decomposition import NMF
from sklearn.metrics.pairwise import cosine_similarity
# gensim
from gensim import corpora, models, similarities, matutils

# sklearn
from sklearn import datasets
from sklearn.feature_extraction.text import CountVectorizer

# logging for gensim (set to INFO)
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
import pickle
import pandas as pd
```

```python
df = pd.read_pickle("./tokenized.pkl")
```

```python
df.head()
```

```python
count_vectorizer = CountVectorizer(ngram_range=(1, 2),  
                                   stop_words='english')

count_vectorizer.fit(df['tokenized'])
```

```python
doc_word = count_vectorizer.transform(df['tokenized']).transpose()
```

```python
pd.DataFrame(doc_word.toarray(), count_vectorizer.get_feature_names()).head()
```

```python
doc_word.shape
```

```python
corpus = matutils.Sparse2Corpus(doc_word)
```

```python
id2word = dict((v, k) for k, v in count_vectorizer.vocabulary_.items()) #dictionary of words 
```

```python
lda = models.LdaModel(corpus=corpus, num_topics=20, id2word=id2word, passes=10)
```

```python
lda.print_topics()
```
